# splitting the predict column from the actual dataset 

```python
x=df.drop("Outcome",axis=1)
x.head()
```

```python
y=df.Outcome
y.head()
```

# splitting into train and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

# training of the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
```

```python
test_accuracy=[]
from sklearn.metrics import confusion_matrix,accuracy_score, precision_score,recall_score,f1_score
from sklearn.metrics import roc_curve,auc
for k in range(1,50,3):
    print("K=",k)
    neigh=KNeighborsClassifier(n_neighbors=k)
    neigh.fit(x_train,y_train)
    print("train confusion matrix")
    print(confusion_matrix(y_train,neigh.predict(x_train)))
    print("test confusion matrix")
    print(confusion_matrix(y_test,neigh.predict(x_test)))
    print(accuracy_score(y_test,neigh.predict(x_test)))
    print("*"*50)
    test_accuracy.append(accuracy_score(y_test,neigh.predict(x_test)))
    
```

```python
k=np.arange(1,50,3)
plt.plot(k,test_accuracy,label='test accuracy')

plt.legend()
plt.xlabel("K:hyperparameter")
plt.ylabel("test accurcay")
plt.title("ERROR PLOTS")
plt.show()
```

```python
neigh=KNeighborsClassifier(n_neighbors=7)
neigh.fit(x_train,y_train)
print("test confusion matrix")
print(confusion_matrix(y_test,neigh.predict(x_test)))
```

# Accuracy, precision, recall, f1 score

```python
#accuracy calculates the correctly predicted observation among the total observation
pred=neigh.predict(x_test)
accuracy=accuracy_score(y_test,pred)
print("accuracy: ",accuracy)
```

```python
#precision calculates out of the predicted positive, how many of them are actually positive
precision=precision_score(y_test,pred)
print("precision: ",precision)
```

```python
#recall calculates out of the actual positive, how many of them were predicted
recall=recall_score(y_test,pred)
print("recall: ",recall)
```

```python
#weighted average of presicion and recall
f1=f1_score(y_test,pred)
print("f1 score: ",f1)
```
